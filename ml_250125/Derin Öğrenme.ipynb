* **Yapay Sinir A Nedir?** 
    * Yapay sinir alar, insan beyninin bilgi ileme tekniinden esinlenerek gelitirilmi bir bilgi ilem teknolojisidir.

    * Lojistik regresyonda, sadece girdi ve kt katmanlar vardr. Ancak sinir anda giri ve k katman arasnda en az bir gizli katman vardr.
    * Yapay sinir a zamanda derin sinir alar olarak da adlandrlrlar.

    * Gizli katmanlar girdileri gremezler. Girdileri gren katman "girdi katman" olarak adlandrlr.
    * Katman says ifade edilirken, girdi katman saylmaz. rnein bir derin renme modeli tasarladnz dnn, ierisinde 1 girdi, 1 gizli ve 1 tane de kt katman olsun. Ka tane katmana sahip bir model tasarladnz sorusunun cevabi 2'dir (1 gizli + 1 kt katmanlar).

#### Parametrelerin lklendirilmesi
* X eitim veri seti iin $x^{(7493)}$:
$$z^{[1] (7493)} =  W^{[1]} x^{(7493)} + b^{[1] (7493)}$$ 
$$a^{[1] (7493)} = \tanh(z^{[1] (7493)})$$
$$z^{[2] (7493)} = W^{[2]} a^{[1] (7493)} + b^{[2] (7493)}$$
$$\hat{y}^{(7493)} = a^{[2] (7493)} = \sigma(z^{ [2] (7493)})$$
* Lojistik regresyonda, 0.01 arlklar ve 0 eilimi ile ilklendirme ilemini gerekletirmitik. Bu sefer, arlklar rastgele olarak ilklendireceiz. nk parametreleri sfrlarsak, ilk gizli katmandaki her sinir ayn karlatrmay gerekletirecektir. Bu nedenle, gradyan iniinin oklu yinelemesinden sonra bile katmandaki her sinir dier sinirlerle ayn parametreleri hesaplayacaktr. Bu nedenle rastgele balatyoruz. 
* Ayrca balang ilklendirilen arlklar sfra yakn seiyoruz. Balangta ok byk deerler seersek, bu durum tanh girdilerinin ok byk olmasna ve dolaysyla gradyanlarn sfra yakn olmasna neden olacaktr. Sonu olarak optimizasyon algoritmas yava olacaktr.
* Eilim balangta sfr olabilir.

```python
# ilklendirme
def layerSayisiniParametreleriIlklendir(x_train, y_train):
    parameters = {"weight1": np.random.randn(3,x_train.shape[0]) * 0.1,
                  "bias1": np.zeros((3,1)),
                  "weight2": np.random.randn(y_train.shape[0],3) * 0.1,
                  "bias2": np.zeros((y_train.shape[0],1))}
    return parameters
```

```python
# sigmoid
def sigmoid(z):
    
    y_head = 1/(1+ np.exp(-z))
    return y_head
sigmoid(0)
```

<a id="11013"></a> <br>
#### leri Yaylma
* leri yaylma, lojistik regresyon ile hemen hemen ayndr.
* Tek fark, tanh fonksiyonunu kullanmamz ve tm ilemleri iki kez yapyor olmamz.
* Ayrca Numpy ktphanesinin tanh fonksiyonu bulunmaktadr. Bu nedenle sigmoid fonksiyonu gibi bir tanh fonksiyonu tanmlamamza gerek bulunmamaktadr.

```python

def ileriYayilmaNN(x_train, parameters):

    Z1 = np.dot(parameters["weight1"],x_train) +parameters["bias1"]
    A1 = np.tanh(Z1)
    Z2 = np.dot(parameters["weight2"],A1) + parameters["bias2"]
    A2 = sigmoid(Z2)

    cache = {"Z1": Z1,
             "A1": A1,
             "Z2": Z2,
             "A2": A2}
    
    return A2, cache
```

<a id="11014"></a> <br>
#### Kayp ve Maliyet Fonksiyonlar
* Kayp ve maliyet fonksiyonlar lojistik regresyon ile ayndr
* apraz entropi fonksiyonu 
* 

```python
# maliyet hesab
def maliyetHesaplaNN(A2, Y, parameters):
    logprobs = np.multiply(np.log(A2),Y)
    cost = -np.sum(logprobs)/Y.shape[1]
    return cost
```

<a id="11015"></a> <br>
#### Geriye Yaylma
* Lojistik regresyondan bildiiniz gibi geriye doru yaylma blmnde maliyetin parametrelere gre trevini alp deiimi buluyoruz.

```python
# Backward Propagation
def geriyeYayilmaNN(parameters, cache, X, Y):

    dZ2 = cache["A2"]-Y
    dW2 = np.dot(dZ2,cache["A1"].T)/X.shape[1]
    db2 = np.sum(dZ2,axis =1,keepdims=True)/X.shape[1]
    dZ1 = np.dot(parameters["weight2"].T,dZ2)*(1 - np.power(cache["A1"], 2))
    dW1 = np.dot(dZ1,X.T)/X.shape[1]
    db1 = np.sum(dZ1,axis =1,keepdims=True)/X.shape[1]
    grads = {"dweight1": dW1,
             "dbias1": db1,
             "dweight2": dW2,
             "dbias2": db2}
    return grads
```

<a id="11016"></a> <br>
#### Parametrelerin Gncellenmesi
* Arlk ve eilim parametrelerimizi gncelleyelim.

```python
# update parameters
def parametreGuncelleNN(parameters, grads, learning_rate = 0.01):
    parameters = {"weight1": parameters["weight1"]-learning_rate*grads["dweight1"],
                  "bias1": parameters["bias1"]-learning_rate*grads["dbias1"],
                  "weight2": parameters["weight2"]-learning_rate*grads["dweight2"],
                  "bias2": parameters["bias2"]-learning_rate*grads["dbias2"]}
    
    return parameters
```

<a id="11017"></a> <br>
#### Tahmin
* Tahmin ilemini gerekletirelim.

```python
# tahmin
def tahminNN(parameters,x_test):
    
    # x test ileri yayilma girdisidir.
    A2, cache = ileriYayilmaNN(x_test,parameters)
    Y_prediction = np.zeros((1,x_test.shape[1]))
    
    # eer z 0.5'den bykse, 1 snfna ait resimler
    # eer z 0.5'den kkse, 0 snfna ait resimler
    for i in range(A2.shape[1]):
        if A2[0,i]<= 0.5:
            Y_prediction[0,i] = 0
        else:
            Y_prediction[0,i] = 1

    return Y_prediction
```

#### 2 Katmanl Sinir A Oluturma
* imdiye kadar yazdmz fonksiyonlar birletirelim.

```python
# 2 - katmanl sinir a
def ikiKatmanNN(x_train, y_train,x_test,y_test, num_iterations):
    cost_list = []
    index_list = []
    
    # ilklendirme
    parameters = layerSayisiniParametreleriIlklendir(x_train, y_train)

    for i in range(0, num_iterations):
         # ileri yaylm
        A2, cache = ileriYayilmaNN(x_train,parameters)
        # maliyet hesapla
        cost = maliyetHesaplaNN(A2, y_train, parameters)
         # geriye yaylm
        grads = geriyeYayilmaNN(parameters, cache, x_train, y_train)
         # parametreleri gncelle
        parameters = parametreGuncelleNN(parameters, grads)
        
        if i % 100 == 0:
            cost_list.append(cost)
            index_list.append(i)
            print ("terasyon: %i, Maliyet: %f" %(i, cost))
            
    plt.plot(index_list,cost_list)
    plt.xticks(index_list,rotation='vertical')
    plt.xlabel("terasyon Sayisi")
    plt.ylabel("Maliyet")
    plt.show()
    
    # predict
    y_prediction_test = tahminNN(parameters,x_test)
    y_prediction_train = tahminNN(parameters,x_train)

    # eitim ve test doruluu
    print("Eitim Doruluu: {} %".format(100 - np.mean(np.abs(y_prediction_train - y_train)) * 100))
    print("Test Doruluu: {} %".format(100 - np.mean(np.abs(y_prediction_test - y_test)) * 100))
    return parameters

parameters = ikiKatmanNN(x_egitim, y_egitim,x_test,y_test, num_iterations=1000)
```

#### Keras Ktphanesi ile Yapay Sinir Alar
Keras kitaplnn baz parametrelerine bakalm:
* units: nron says
* kernel_initializer: arlklar ilklendirme
* activation: aktivasyon fonksiyonu
* input_dim: input boyutu bizim rneimizde piksel says (784)
* optimizer: Adam optimizasyon yntemini reneceiz.
    * Adam, sinir alarn eitmek iin en etkili optimizasyon algoritmalarndan biridir.
    * Adam'n baz avantajlar, nispeten dk bellek gereksinimlerinin olmas ve genellikle kk hiperparametrelerin ayarlanmasyla bile iyi almasdr.
* loss: apraz entropi maliyet fonksiyonu kullanacaz.
$$J = - \frac{1}{m} \sum\limits_{i = 0}^{m} \large\left(\small y^{(i)}\log\left(a^{[2] (i)}\right) + (1-y^{(i)})\log\left(1- a^{[2] (i)}\right)  \large  \right) \small \tag{6}$$
* metrics: metrik olarak doruluu kullanacaz
* cross_val_score: apraz dorulama kullanacaz.
* epochs: iterasyon says

```python
# yapay sinir alar ktphaneleri
from keras.wrappers.scikit_learn import KerasClassifier
from sklearn.model_selection import cross_val_score
from keras.models import Sequential # initialize neural network library
from keras.layers import Dense # build our layers library

# modelimiz iin bir metot yazalm
def siniflandiriciMethot():
    classifier = Sequential() # initialize neural network
    classifier.add(Dense(units = 8, kernel_initializer = 'uniform', activation = 'relu', input_dim = x_egitim.shape[1]))
    classifier.add(Dense(units = 4, kernel_initializer = 'uniform', activation = 'relu'))
    classifier.add(Dense(units = 1, kernel_initializer = 'uniform', activation = 'sigmoid'))
    classifier.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
    return classifier

siniflandirici = KerasClassifier(build_fn = siniflandiriciMethot, epochs = 5)
dogruluk = cross_val_score(estimator = siniflandirici, X = x_egitim, y = y_egitim, cv = 3)# 3l arpraz dorulama
ortalama = dogruluk.mean()
varyans = dogruluk.std()
print("Doruluun ortalamas: "+ str(ortalama))
print("Doruluun Varyans: "+ str(varyans))
```

<a id="1115"></a>
#### Evriim Operasyonu Nedir?
* ncelikli olarak bir grnt ve zellik alglaycmz (3 * 3) olsun.
* zellik dedektrnn 3'e 3 matris olmas gerekmez. 5'e 5 veya 7'ye 7 olabilir.
* zellik alglayc ekirdek ve ya filtre alglayc olarak da adlandrlabilir.
* zellik alglayc, kenarlar veya dbkey ekiller gibi zellikleri alglar. rnein, grntmz bir kpek ise, zellik alglayc kpein kula veya kuyruu gibi zellikleri alglayabilir.
* zellik haritas = evriim (giri resmi, zellik alglayc). Matrislerin elemanlarnn arpm.
* Grnt zerinde gezinme gerekleir.
* Bu ilemler sonucunda orjinal grntnn boyutu azalm oluyor, bu modelimizin hzl almas asndan nemli.
* Birden ok zellik haritas oluturuyoruz, nk birden ok zellik alglayc (filtre) kullanyoruz.
* 
* Evriim katmanndan sonra aktivasyon fonksiyonu olarak ReLU kullanyoruz. Bu aktivasyonu fonksiyonu dorusall krarak, modelimizin dorusal olmayan yaplar renmesine olanak salyor.
*

### Keras ile Evriimsel Sinir Alar

<a id="11111"></a>
#### Model Oluturma
* Keras ktphanesi ile oluturacamz modelde srasyla aada bulunan katmanlar olacaktr.
    * conv => max pool => dropout => conv => max pool => dropout => fully connected (2 layer)
* Dropout: Seyreltme, rastgele seilen nronlarn eitim srasnda gz ard edildii bir tekniktir. Ezberlemeyi(over-fitting) nler.
* 

```python
# ktphaneleri ykleyelim
from sklearn.metrics import confusion_matrix
import itertools
from keras.utils.np_utils import to_categorical # convert to one-hot-encoding
from keras.models import Sequential
from keras.layers import Dense, Dropout, Flatten, Conv2D, MaxPool2D
from keras.optimizers import RMSprop,Adam
from keras.preprocessing.image import ImageDataGenerator
from keras.callbacks import ReduceLROnPlateau

model = Sequential()
#
model.add(Conv2D(filters = 8, kernel_size = (5,5),padding = 'Same', 
                 activation ='relu', input_shape = (28,28,1)))
model.add(MaxPool2D(pool_size=(2,2)))
model.add(Dropout(0.25))
#
model.add(Conv2D(filters = 16, kernel_size = (3,3),padding = 'Same', 
                 activation ='relu'))
model.add(MaxPool2D(pool_size=(2,2), strides=(2,2)))
model.add(Dropout(0.25))
# tam balant
model.add(Flatten())
model.add(Dense(256, activation = "relu"))
model.add(Dropout(0.5))
model.add(Dense(10, activation = "softmax"))
```

<a id="11112"></a>
#### Optimizerin Belirlenmesi 
* Adam optimizasyon yntemini kullanacaz.

```python
# Adam optimizasyon yntemini kullanacaz.
optimizer = Adam(lr=0.001, beta_1=0.9, beta_2=0.999)
```

<a id="11113"></a>
#### Modelin Derlenmesi
* Kategorik apraz entropi maliyet yntemini kullanacaz.
* 


```python
# modeli derleyelim
model.compile(optimizer = optimizer , loss = "categorical_crossentropy", metrics=["accuracy"])
```

#### Dng Says ve Eitim Boyutu
* 10 rnek ieren bir veri kmesi olduunu varsayalm. 
* batch_size = 2 olsun
* epochs = 3 olsun.
* Her bir epoch iin 5 adet batch'e sahibiz (10/2 = 5).
* 10 adet veri, her epoch da 2'erli olarak 5 kerede dngsn tamamlar.

```python
epochs = 10  
batch_size = 250
```

* Ezberleme sorununu nlemek iin, el yazs rakam veri setimizi yapay olarak geniletmemiz gerekiyor.
* Rakam varyasyonlarn yeniden oluturmak iin eitim verilerini kk dnmlerle deitirebiliriz.

```python
# data augmentation
datagen = ImageDataGenerator(
        featurewise_center=False,  # girdi ortalamasn veri kmesi zerinden 0 olarak ayarlayn
        samplewise_center=False,  # her bir rnek ortalamasn 0 olarak ayarlayn
        featurewise_std_normalization=False,  # girdileri veri kmesinin standartlarna bl
        samplewise_std_normalization=False,  # her girdiyi std'ye bl
        zca_whitening=False,  # dimesion reduction
        rotation_range=5,  # boyut kltme
        zoom_range = 0.1, # % 10 grnty rastgele yaknlatr
        width_shift_range=0.1,  # grntleri yatay olarak rastgele kaydr % 10
        height_shift_range=0.1,  # grntleri dikey olarak rastgele kaydr % 10
        horizontal_flip=False,  # grntleri yatay olarak rastgele evir
        vertical_flip=False)  # grntleri yadikeytay olarak rastgele evir 

datagen.fit(X_egitim)
```

```python
# modelin eitilmesi
history = model.fit_generator(datagen.flow(X_egitim,Y_egitim, batch_size=batch_size),
                              epochs = epochs, validation_data = (X_test,Y_test), steps_per_epoch=X_egitim.shape[0] // batch_size, verbose=2)
```

```python
# veri setlerini ykleyelim
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
```

```python
# veriyi ie aktar
veri = pd.read_csv('uluslararas_havayolu_yolcular.csv')
veri.head()
```

```python
# veriyi grselletirelim
veri = veri.iloc[:,1].values
plt.plot(veri)
plt.xlabel("zaman")
plt.ylabel("Yolcu Says")
plt.title("Uluslararas Havayolu Yolcular")
plt.show()
```

```python
# verinin eklini ve veri tipini deitir
veri = veri.reshape(-1,1)
veri = veri.astype("float32")
veri.shape
```

```python
# normalize edelim
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler(feature_range = (0, 1))
veri_scaled = scaler.fit_transform(veri)
veri_scaled[:4]
```

```python
# veriyi eitim ve test olarak blelim,
# verinin %60'lk ilk ksm eitim geri kalan da test olsun
egitim_boyu = int(len(veri) * 0.60)
test_boyu = len(veri) - egitim_boyu
egitim =veri_scaled[0:egitim_boyu,:]
test = veri_scaled[egitim_boyu:len(veri),:]
print("Egitim boyu: {}, Test boyu: {} ".format(len(egitim), len(test)))
```

```python
# Eitim verisini 10'ar l olarak yeniden ekillendirelim
time_stemp = 10
dataX = []
dataY = []
for i in range(len(egitim)-time_stemp-1):
    a = egitim[i:(i+time_stemp), 0]
    dataX.append(a)
    dataY.append(egitim[i + time_stemp, 0])
egitimX = np.array(dataX)
egitimY = np.array(dataY)  
```

```python
# test verisini 10'ar l olarak yeniden ekillendirelim
dataX = []
dataY = []
for i in range(len(test)-time_stemp-1):
    a = test[i:(i+time_stemp), 0]
    dataX.append(a)
    dataY.append(test[i + time_stemp, 0])
testX = np.array(dataX)
testY = np.array(dataY) 
```

```python
# verinin eklini keras iin uygun hale getirelim
egitimX = np.reshape(egitimX, (egitimX.shape[0], egitimX.shape[1], 1))
testX = np.reshape(testX, (testX.shape[0], testX.shape[1], 1))
print('Eitim X ekli: ', egitimX.shape)
print('Test X ekli: ', testX.shape)
```

<a id="11232"></a>
#### Create RNN Model

```python
# keras ktphanemizde gerekli metotlar ie aktaralm
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import SimpleRNN
from keras.layers import Dropout

# tekrarlanan sinir alarnn temelini oluturur.
regressor = Sequential()

# ilk katman ve drop out katman
regressor.add(SimpleRNN(units = 120,activation='relu', return_sequences = True, input_shape = (egitimX.shape[1], 1)))
regressor.add(Dropout(0.2))

# ikinci katman ve drop out katman
regressor.add(SimpleRNN(units = 80,activation='relu', return_sequences = True))
regressor.add(Dropout(0.2))

# nc katman ve drop out katman
regressor.add(SimpleRNN(units = 40,activation='relu', return_sequences = True))
regressor.add(Dropout(0.2))

# drdnc katman ve drop out katman
regressor.add(SimpleRNN(units = 20,activation='relu', return_sequences = True))
regressor.add(Dropout(0.2))

# beinci katman ve drop out katman
regressor.add(SimpleRNN(units = 10))
regressor.add(Dropout(0.2))

# tek bir kt olacak 
regressor.add(Dense(units = 1))

# derleyelim
regressor.compile(optimizer = 'adam', loss = 'mean_squared_error')

# eitim balasn
regressor.fit(egitimX, egitimY, epochs = 250, batch_size = 50, verbose = 0)
```

<a id="11233"></a>
#### Tahminleme ve Modelin Deerlendirilmesi
* Modelin deerlendirilmesinide kk-ortalama-kare hatas yntemini kullanacaz. 
* 

```python
# gerekli ktphaneleri ekleyelim.
import math 
from sklearn.metrics import mean_squared_error

egitimTahmin = regressor.predict(egitimX)
testTahmin = regressor.predict(testX)

# normalize etmitik normale evirelim
egitimTahmin = scaler.inverse_transform(egitimTahmin)
egitimY = scaler.inverse_transform([egitimY])
testTahmin = scaler.inverse_transform(testTahmin)
testY = scaler.inverse_transform([testY])
```

```python
# kk ortalama kara hatas
trainScore = math.sqrt(mean_squared_error(egitimY[0], egitimTahmin[:,0]))
print('Eitim Skoru: %.2f RMSE' % (trainScore))
testScore = math.sqrt(mean_squared_error(testY[0], testTahmin[:,0]))
print('Test Skoru: %.2f RMSE' % (testScore))
```
